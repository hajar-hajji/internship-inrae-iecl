### Extraire les composantes (annuelles et cycliques) à partir de GPP

$$X_t =
\overline{X}^{365d}_t +
\overline{X_t - \bar{X}^{365d}_t} +
\tilde{X_t}$$

avec :
$$\overline{X}^{T} = \frac{1}{N} \sum_{k=1}^{N} X_{t+kT}$$

```python
from datetime import datetime, timedelta

# calculer la composante annuelle de GPP en choisissant N=2
T=365
interval_T = timedelta(days=T)
interval_2T = timedelta(days=2*T)

# calcul de X_t pr chaque t
for idx, row in df.iterrows():
    date_t = row['DateTime']
    date_t_plus_T = date_t + interval_T
    date_t_plus_2T = date_t + interval_2T

    # recherche des observ corresp pour t+T et t+2*T
    X_t_plus_T = df[df['DateTime'] == date_t_plus_T]['GPP'].values
    X_t_plus_2T = df[df['DateTime'] == date_t_plus_2T]['GPP'].values

    if len(X_t_plus_T) > 0 and len(X_t_plus_2T) > 0:
        df.loc[idx, 'GPP_ann'] = 0.5*(X_t_plus_T[0] + X_t_plus_2T[0])
```

```python
df.head()
```

```python
df['GPP_t - GPP_ann'] = df['GPP'] - df['GPP_ann']
df.head()
```

```python
# idem pr calculer la composante qui décrit les cycles, N=7
T=1
interval_T = timedelta(days=T)

# calcul de X_t pr chaque t
for idx, row in df.iterrows():
    date_t = row['DateTime']
    s=0
    for i in range(1,8):
      date_t_plus_iT = date_t + i*interval_T

      # recherche des observ corresp pour t+T jusqu'à t+7T
      X_t_plus_iT = df[df['DateTime'] == date_t_plus_iT]['GPP_t - GPP_ann'].values

      if len(X_t_plus_iT) > 0:
        s += X_t_plus_iT

    df.loc[idx, 'GPP_cycle'] = s/7
```

```python
df['anomalies'] = df['GPP'] - df['GPP_ann'] - df['GPP_cycle']
df.head()
```

### Configurer le temps

```python
df['Time (s)'] = df['DateTime'].apply(lambda x: x.timestamp()) # convertir t en secondes
df['Time (s)'] = df['Time (s)'] - df['Time (s)'][0] # commencer de 0
df.head()
```

```python
# on a privilégié une forme sinusoïdale pour le temps (entrée utile pour le modèle, celui-ci saura se situer dans le temps)
d = 24*3600
df['cos day'] = np.cos(df['Time (s)']*(2*np.pi/d))
df['sin day'] = np.sin(df['Time (s)']*(2*np.pi/d))

y = 365*d
df['cos year'] = np.cos(df['Time (s)']*(2*np.pi/y))
df['sin year'] = np.sin(df['Time (s)']*(2*np.pi/y))
df.head()
```

```python
# récupérer la dernière date
last_date = df.iloc[-1].DateTime
my_interval = pd.DateOffset(years=2, days=7) # intervalle de 2 ans et 7 jours

# on élimine les observations prises dans les 2 dernières années et 7 jours car on a pas pu calculer leurs composantes de GPP
date_elimination = last_date - my_interval
final_df = df[df['DateTime'] <= date_elimination]
```

```python
final_df = final_df.drop('GPP_t - GPP_ann', axis=1)
```

```python
final_df.isna().sum() # plus de composantes NaN pour les comp annuelles, cycliques ainsi que les anomalies
```
